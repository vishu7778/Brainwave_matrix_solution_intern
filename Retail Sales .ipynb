```python
#Ploting a chart to understand quickly
ax = sns.countplot(x= 'Gender',data= data)
for bars in ax.containers:
     ax.bar_labels(bars)
    
```

```python
#now plottig a bar chart for gender vs total amount
sales_gender= data.groupby(['Gender'], as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
sns.barplot(x='Gender', y='Quantity', data=sales_gender)
```

```python
ax= sns.countplot(data= data, x='Quantity', hue= 'Gender')
for bars in ax.containers:
    ax.bar_labels(bars)
```
